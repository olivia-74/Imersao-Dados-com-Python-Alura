```python
rename_column = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}

df.rename(columns=rename_column, inplace=True)

display(df.head())
```

```python
senioridade = {
    'EN': 'Junior',
    'MI': 'Pleno',
    'SE': 'Senior',
    'EX': 'Executivo'
}

df['senioridade'] = df['senioridade'].replace(senioridade)

df['senioridade'].value_counts()
```

```python
contrato = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'FL': 'Freelancer',
    'PT': 'Tempo Parcial'
}

df['contrato'] = df['contrato'].replace(contrato)

df['contrato'].value_counts()
```

```python
tamanho_empresa = {
    'S': 'Pequena',
    'M': 'Média',
    'L': 'Grande'
}

df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa)

df['tamanho_empresa'].value_counts()
```

```python
remoto = {
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
}

df['remoto'] = df['remoto'].replace(remoto)

df['remoto'].value_counts()
```

```python
# criando um dataframe de exemplo
import numpy as np # biblioteca numerica

df_salarios = pd.DataFrame({
  'nome': ['olivia', 'felipe', 'tofu', 'shoyu', 'gatao'],
  'salario': [10000, np.nan, 3000, np.nan, 200000]
})


# nova coluna salario_media
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))
# fillna: preencher os valores nulos de df_salario em salario com a média, arredondando para duas casas decimais

# nova coluna salario_mediana
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

display(df_salarios)
```

```python
df_temperaturas = pd.DataFrame({
  'dia': ['segunda', 'terça', 'quarta', 'quinta', 'sexta'], #lista
  'temperatura': [30, np.nan, np.nan, 28, 27]
})

df_temperaturas['preenchido_ffill'] = df_temperaturas['temperatura'].ffill()
# ffill: preenche os valores nulos com o proximo valor, forward fill
df_temperaturas['preenchido_bfill'] = df_temperaturas['temperatura'].bfill()
# ffill: preenche os valores nulos com o valor anterior, backward fill

df_temperaturas
```

```python
# criação de lista usa []
df_cidades = pd.DataFrame({
  'nome': ['olivia', 'felipe', 'tofu', 'shoyu', 'gatao'],
  'cidade': ['São Paulo', np.nan, 'São Paulo', np.nan, 'Rio de Janeiro']
})

df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('não informado')
df_cidades
```

```python
# removendo os valores null
df_limpo = df.dropna()
df_limpo.isnull().sum()
df_limpo.head()
df_limpo.info() #carrega Column Non-Null Count Dtype
```

```python
# alterando o tipo da variavel ano para inteiro
df_limpo = df_limpo.assign(ano = df_limpo['ano'].astype('int64'))
df_limpo.info()
```

```python
import plotly.express as px

# Calculate the average salary by seniority
df_senioridade_salario_medio = df_limpo.groupby('senioridade')['usd'].mean().reset_index()

# Define the order of seniority levels
ordem_senioridade = ['Junior', 'Pleno', 'Senior', 'Executivo']

# Create the bar chart
fig = px.bar(df_senioridade_salario_medio,
             x='senioridade',
             y='usd',
             category_orders={'senioridade': ordem_senioridade},
             labels = {'senioridade': 'nivel de senioridade', 'usd':'media salarial anual'},
             title='Média Salarial por Senioridade (USD)')

# Show the plot
fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index() #conta frequencia]
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names= 'tipo_trabalho',
             values= 'quantidade',
             title='proporção dos tipo de trabalho',
             hole=0.5
             )


fig.show()
```

```python
# adiciona porcentagem

remoto_contagem = df_limpo['remoto'].value_counts().reset_index() #conta frequencia]
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names= 'tipo_trabalho',
             values= 'quantidade',
             title='proporção dos tipo de trabalho',
             hole=0.5
             )

fig.update_traces(textinfo='percent+label')
fig.show()
```

```python
cargo_cientista_dados = df_limpo[df_limpo['cargo'] == 'Data Scientist']  # dataframe para o cargo de cientista de dados
salario = cargo_cientista_dados.groupby('residencia')['usd'].mean().reset_index()
fig = px.bar(salario, # Use the aggregated dataframe
             x='residencia', # Use the correct column name
             y='usd',
             labels = {'residencia': 'pais', 'usd':'media salarial anual'},
             title='Media salarial de um cientista de dados por pais')
fig.update_layout(xaxis={'categoryorder':'total ascending'}) # ordena por salario


fig.show()
```
